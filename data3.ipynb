```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("amazon_product.csv")
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1.columns
```

```python
df1.info()
```

```python
""" 
changing the columns name type  
deleting redundant columns 
renaming the columns 
dropping duplicates
remove tha nan  value from the dataset
check for some more transformation   

"""
```

```python
#cleaning colummns names  
df1.columns = (
    df1.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df1.columns
```

```python
#capitalise the columns name 
new_columns = []

for i in df1.columns:
    new_columns.append(i.capitalize())
```

```python
df1.columns = new_columns
```

```python
df1.columns
```

```python
#dropiing the unnecessary columns 
df1.drop(columns = [ 'Unnamed:_0', 'Asin','Product_url','Product_photo','Product_num_offers','Product_minimum_offer_price','Is_amazon_choice', 'Is_prime', 'Climate_pledge_friendly',
       'Sales_volume', 'Delivery', 'Has_variations'], inplace = True)
```

```python
df1.columns 
```

```python
df1
```

```python
#droppping duplicates daata 
df1.duplicated().sum()
```

```python
#whwat if we had duplicated 
df1.drop_duplicates(inplace = True)
```

```python
#for checking NAN values 
df1.isna().sum()
```

```python
df1.shape
```

```python
df1 = df1[["Product_title","Product_price","Currency","Product_star_rating","Product_num_ratings","Is_best_seller"]]
```

```python
df1
```

```python
df1.isna().sum()
```

```python
#dropping na values
df1.dropna(inplace = True)
```

```python
df1.isna().sum()
```

```python
#some more transformation
df1.head(3)
```

```python
df1["Is_best_seller"].unique()
```

```python
#changing the values into boolean form 
df1["Is_best_seller"] = df1["Is_best_seller"].apply(lambda x:1 if x == True else 0)
```

```python
#for checking the columns and sign
df1["Product_price"]
```

```python
df1["Product_price"] = df1["Product_price"].str.replace("$","")
```

```python
#for changing the data type   of values 
df1["Product_price"] = df1["Product_price"].astype(float)
```

```python
df1
```

```python
df1.info()
```

```python
#product price dsistribution 
sns.histplot(df1['Product_price'], bins = 12, kde= True)
plt.title("Distribution of product prices")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

```python
#star rating Distribution

sns.countplot(
    x='Product_star_rating', 
    data = df1
)

plt.title("Distribution of product rating")
plt.xlabel("Star Rating")
plt.ylabel("Numbers of product")
```

```python
#bestsseller count
sns.countplot(
    x='Is_best_seller',
    data=df1
)

plt.title("Bestseller vs Non best seller")
plt.xlabel("Is best seller")
plt.ylabel("count")

plt.show()
```

```python
#rating vs Number of rating

sns.scatterplot(
    x='Product_star_rating',
    y= "Product_num_ratings",
    hue='Is_best_seller',
    data = df1,
    s = 120
)

plt.title("Rating vs number of rating")
plt.xlabel("Star Rating")
plt.ylabel("Number of rating")
plt.show()
```

```python
#Aveerage price by Bestseller
sns.barplot(
    x='Is_best_seller',
    y='Product_price',
    data = df1
)

plt.title("Average price by Bestseller status")
plt.xlabel("Best seller")
plt.ylabel("Average price")

plt.show()
```

```python
#top 10 most Reviewed Product 
top_product = df1.sort_values(
    by='Product_num_ratings',
    ascending = False
).head(10)


sns.barplot(
    x='Product_num_ratings',
    y='Product_title', 
    data = top_product
)

plt.title("Top 10 most reviewed products ")
plt.xlabel("Nummber of rating")
plt.ylabel("product Title")

plt.show()
```

```python
#correlation heatmap 
corr = df1[
    [
        'Product_price',
        'Product_star_rating',
        'Product_num_ratings'
    ]
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("correlation Heatmap")
plt.show()
```

```python
#Boxplot for price analysis 

sns.boxplot(
    x='Is_best_seller',
    y = 'Product_price',
    data = df1
)

plt.title("Price distribution by bestseller")

plt.show()
```
